### Data Cleaning 
<br> 1. **df_gun needs to have column attributes to be all lowercase with no spaces**
<br>
<br> 2. **df_gun needs to seperate the data in column "month" to "month" and "year"**
<br>
<br> 3. **change "multiple" and "totals" in df_gun both to floats for consistency and year to an int**
<br>
<br> 4. **delete "fact note" column in df_census since it not useful**
<br>
<br> 5. **delete extra rows on the bottom of df_census that contain notes**
<br>
<br> 6. **flip columns and rows in df_census**
<br>
<br> 7. **df_census needs to fix column attribute names to all lowercase with no spaces**
<br>
<br> 8. **adjust column names in df_census to be more readable and useful**
<br> 
<br> 9. **change columns in df_census to floats**

```python
# 1. make all the index of the columns of df_gun lowercase with no spaces
df_gun.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
df_gun.head(1)
```

```python
# 2. seperate "month" to "month" and "year" in df gun
df_gun[['month','real_month']] = df_gun['month'].str.split('-', expand = True)
df_gun.head()
```

```python
# rename 'month' -> 'year' and 'real_month' -> 'month'
df_gun.rename(columns = {'month':'year'}, inplace = True)
df_gun.rename(columns = {'real_month' : 'month'}, inplace = True)
df_gun.head()
```

```python
# put "month" as the first column
first_col = df_gun['month']
df_gun.drop(labels = ['month'], axis = 1, inplace= True)
df_gun.insert(0, 'month', first_col)
df_gun.head()
```

```python
# 3. change "multiple" and "totals" in df_gun both to floats for consistency, year -> int
df_gun['multiple'] = df_gun['multiple'].astype(float)
df_gun['totals'] = df_gun['totals'].astype(float)
df_gun['year'] = df_gun['year'].astype(int)
df_gun.dtypes
df_gun.head()
```

```python
# 4. delete factnote column from the census
df_census.drop(columns=['Fact Note'], inplace=True)
df_census.head(1)
```

```python
# 5. delete rows of notes on the bottom
df_census.tail(10)
df_census.drop(df_census.index[64:85], inplace=True)
df_census.tail(10)
```

```python
# 6. flip columns and rows in df_census to cast columns as floats (and adjust names accordingly for percentages)
df_census = df_census.transpose()
df_census.head()
```

```python
# 7. df_census needs to fix column attribute names to all lowercase with no spaces
df_census.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
df_census.rename(columns=lambda x: x.strip().replace("'", ""), inplace = True)
df_census.head(2)
```

```python
# 8. adjust column names in df_census to be more readable and useful
# take out "," and "(v2016)"
for col in df_census.columns:
    print(col)
```

```python
df_census.rename(columns=lambda x: x.strip().replace("(v2016)", ""), inplace=True)
df_census.rename(columns=lambda x: x.strip().replace(",", ""), inplace = True)
df_census.rename(columns=lambda x: x.strip().replace("__", ""), inplace=True)
#the double underscore would not come off with "(v2016)" so needed a seperate function
df_census.head(2)
```

```python
# 9. change columns in df_census to all floats
#there are some letters that need to be replaced by zeros
df_census = df_census.replace(['FN', 'D', 'Z'], '0')
#every value that had a letter needs to be a zero
df_census.fillna(0, inplace=True)
#now that all the numbers are cleaned I can convert them all to floats
df_census = df_census.astype(float)
```

### What census population data coorelates to the states with the highest gun permits in the 21st century?

```python
#### only selecting the years in df_gun in 2000-2017
df_gun_years = df_gun.query('year > 1999')

#looking at the top 10 states with the highest mean of permits
top = df_gun_years.groupby('state').permit.mean().nlargest(4)
top
```

```python
top = df_gun_years.groupby('state').permit.median().nlargest(4)
top
```

```python
#index column does not have a name so need to change that 
df_census = df_census.rename_axis('state')

#creating new dataframe with only the states of high gun permits
df_states = df_census.loc[["Kentucky", "Illinois", "California", "North Carolina"]]
```

```python
df_census['with_a_disability_under_age_65_years_percent_2011-2015'].mean() * 100
#kentucky is number 2 but I don't see any other high permit states
```

```python
dis = df_states['with_a_disability_under_age_65_years_percent_2011-2015'] * 100
#dis.round(decimals=2)
dis
```

```python
vets = (df_states['veterans_2011-2015']/df_states['population_estimates_july_1_2016']) * 100
vets
#percent of veterans in high gun permit states
```

```python
(df_census['veterans_2011-2015']/df_census['population_estimates_july_1_2016']).mean() *100
#highest vet percent in all states
```

```python
state_labels = ['KY', 'IL', 'CA', 'NC']
x_axis = [i for i, _ in enumerate(dis)]

plt.bar(x_axis, dis, color='purple')
plt.xlabel("State")
plt.ylabel("Percent of Disability")
plt.title("Percentage of Disability Persons in Highest Gun Permit States of the 21st Century")
plt.xticks(x_axis, state_labels)
plt.show()
```

```python
x_axis = [i for i, _ in enumerate(vets)]

plt.bar(x_axis, vets, color='pink')
plt.xlabel("State")
plt.ylabel("Percent of Veterans")
plt.title("Percentage of Veterans in Highest Gun Permit States of the 21st Century")
plt.xticks(x_axis, state_labels)
plt.show()
```

There looks like a coorelation between states with high gum permits, veteran percentage, and disability percentage but North Carolina looks like an outlier in both charts.
